```python
import pandas as pd
import numpy as np

df = pd.read_csv('file1.csv')
df

print(df.isnull().sum())
print(df.dtypes)
```

```python
df
```

DROPPING NA VALUES

```python
df1 = df.dropna()
df1
```

DROPPING THE DUPLICATE VALUES

```python
df2 = df.drop_duplicates()
df2
```

```python
df3= df.isnull()
df3
```

SUBSTITUTING THE NULL VALUES BY DEFAULT METHOD, FOWARD FILL METHOD AND BACKWARD FILL METHOD.

```python
df4=df.fillna(0)
df4
```

```python
df5 =df.fillna(method= 'ffill')
df5
```

```python
df6=df.fillna(method='bfill')
df6
```

```python
#storing the clean data
FILE1=df.to_csv('cleaned_file1.csv', index=False)
```
